```python
import pandas as pd
import numpy as np
from palmerpenguins import load_penguins
import graphviz
from IPython.display import SVG, display
import ipywidgets as widgets
```

```python
penguins_df = load_penguins()
penguins_df.head()
```

```python
penguins_df.drop('year', axis = 1, inplace = True)
penguins_df.dropna(inplace = True)
penguins_df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
penguins_df['species'] = le.fit_transform(penguins_df['species'])
penguins_df['island'] = le.fit_transform(penguins_df['island'])
penguins_df['sex'] = le.fit_transform(penguins_df['sex'])
```

```python
X = penguins_df.drop('species', axis = 1)
y = penguins_df['species']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from graphviz import Source
```

```python
estimator = DecisionTreeClassifier(criterion = 'gini', max_depth = 3)
estimator.fit(X,y)
graph = Source(tree.export_graphviz(estimator
  , out_file=None
  , feature_names=X.columns
  , class_names=['Adelie', 'Chinstrap', 'Gentoo']
  , filled = True))
```

```python
display(SVG(graph.pipe(format = 'svg')))
```

```python
def plot_tree(crit, split, depth, max_feat):
  estimator = DecisionTreeClassifier(criterion = crit, splitter = split, max_depth = depth, max_features = max_feat)
  #clf = DecisionTreeClassifier(max_depth = depth)
  estimator.fit(X,y)
  plt.figure(figsize = (10,8))
  tree.plot_tree(estimator, feature_names = X.columns, class_names = ['Adelie', 'Chinstrap', 'Gentoo'])
  plt.show()
```

```python
grid = widgets.GridspecLayout(35, 35)

grid[8:10, 0:2] = widgets.Button(description = 'Max Depth')
grid[10:13,0:3] = widgets.IntSlider(
    value=7,
    min=0,
    max=10,
    step=1,
    description='',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
    layout= widgets.Layout(width='210px')
)

grid[15:17, 0:3] = widgets.Button(description = 'Split')

grid[17:20,0:3] =widgets.Dropdown(
    options=['best', 'random'],
    value='best',
    description='',
    disabled=False,
)


grid[:,10:33] = plot_tree(crit = grid[10:12,33:-1], split = grid[17:20,0:3], depth = grid[10:13,0:3], max_feat = grid[13:15,33:-1])


grid[10:12,33:-1] = widgets.Dropdown(
    options=['gini', 'impurity'],
    value='gini',
    description='Criterion',
    disabled=False,
)
grid[13:15,33:-1] = widgets.Dropdown(
    options=['auto', 'sqrt', 'log2'],
    value='auto',
    description='Max Features',
    disabled=False,
)
#grid[1:3,1:-1] = widgets.Button(description="Button6", layout=widgets.Layout(height="auto", width="auto"), button_style="info")


```
